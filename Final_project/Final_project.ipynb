```python
from datascience import *
import numpy as np
np.set_printoptions(threshold=50)

import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
students = Table().read_table('https://raw.githubusercontent.com/Matthew-HMS/Datascience101/master/110_sdata.csv')
students
```

```python
# 刪除技職大學
students = students.where('體系別', are.equal_to('1 一般'))
students
```

```python
# 刪除研究所
students = students.where('等級別', are.equal_to('B 學士'))
```

```python
# 只留下日間部
students = students.where('日間∕進修別', are.equal_to('D 日'))
students
```

```python
# 將數字中的逗號去除
adjust = students.column('學生數')

for i in np.arange(len(adjust)):
    adjust[i] = int(adjust[i].replace(',', ''))

students = students.with_column('學生數(改)', adjust)
students
```

```python
# 將資料型態從<U5轉為int32
change = students.column('學生數(改)').astype(np.int32)
students = students.with_column('學生數(改改)', change)
students = students.drop('學生數(改)').drop('學生數').relabeled('學生數(改改)', '學生數')
students
```

```python
#刪除教師數為0的資料並選出學校名稱、科系名稱、學生數、教師數
students = students.where('教師數', are.not_equal_to(0))
students = students.select('學校名稱', '科系名稱', '學生數', '教師數')
```

以下統計資料皆為台灣普通大學(非技職)日間部且是大學部的學生 ! ! !

## 1. 中央大學各學系師生比

```python
ncu_teacher_student = students.where('學校名稱', are.equal_to('國立中央大學'))
ncu_teacher_student
```

```python
ratio = ncu_teacher_student.column('學生數').astype(float) / ncu_teacher_student.column('教師數').astype(float)
ncu_teacher_student_ratio = ncu_teacher_student.with_column('師生比', ratio)
ncu_teacher_student_ratio
```

```python
ncu_teacher_student_ratio.sort('師生比', descending = False).barh('科系名稱', '師生比')
```

此圖是中央大學各學系師生比，可看出管院及學士班的老師對應到的學生數量較多，而其他系所的老師對應到的學生數量較少。

## 2. 全國各大學資管系師生比

```python
mis_teacher_student = students.where('科系名稱', are.equal_to('資訊管理學系'))
mis_teacher_student
```

```python
mis_teacher_student = mis_teacher_student.with_column('師生比', mis_teacher_student.column('學生數').astype(float) / mis_teacher_student.column('教師數').astype(float))
mis_teacher_student
```

```python
mis_teacher_student.sort('師生比', descending = False).barh('學校名稱', '師生比')
```

以上為全國各大學資管系師生比，可看出私立大學資管系的老師對應到的學生數量較多，而國立大學的老師對應到的學生數量較少。

## 3. 全國各大學師生比

```python
group_students_and_teacher_all = students.drop('科系名稱').group('學校名稱', sum)
group_students_and_teacher_all
```

```python
ratio = group_students_and_teacher_all.column('學生數 sum')/ group_students_and_teacher_all.column('教師數 sum')
group_students_and_teacher_all_ratio = group_students_and_teacher_all.with_column('師生比', ratio)
group_students_and_teacher_all_ratio
```

```python
# 全國各大學師生比
group_students_and_teacher_all_ratio.sort('師生比', descending = False).barh('學校名稱', '師生比')
```

```python
#國立(市立)大學平均師生比
gov_ratio_mean = group_students_and_teacher_all_ratio.where('學校名稱',are.containing('立')).column('師生比').mean()
gov_ratio_mean
```

```python
#私立大學平均師生比
own_ratio_mean = group_students_and_teacher_all_ratio.where('學校名稱',are.not_containing('立')).column('師生比').mean()
own_ratio_mean
```

由此可知，平均來說國立大學的師生比較低，而私立大學的師生比較高。

## 4. 推算出台灣所有大學校系平均師生比(Bootstrap)

```python
ratio = students.column('學生數') / students.column('教師數')
students = students.with_column('師生比', ratio)
students
```

```python
# 隨機抽取100筆資料
our_sample = students.sample(100, with_replacement = False)
pop_median = percentile(50, students.column('師生比'))
```

```python
percentile(50, our_sample.column('師生比'))
```

```python
def one_bootstrap_median():
    resample = our_sample.sample(100, with_replacement = True)
    return percentile(50, resample.column('師生比'))
```

```python
# 重複抽樣1000次
repetitions = 1000
bstrap_median = make_array()
for i in np.arange(repetitions):
    bstrap_median = np.append(bstrap_median, one_bootstrap_median())
```

```python
resampled_medians = Table().with_column('Bootstrap Sample Median', bstrap_median)
resampled_medians.hist(bins = np.arange(0, 50, 2))

left = percentile(2.5, bstrap_median)
right = percentile(97.5, bstrap_median)

plt.ylim(-0.01, 0.4)
plt.plot([left, right], [0, 0], color='yellow', lw=3, zorder=1);
plt.scatter(pop_median, 0, color = 'red', s=20, zorder=2);
```
