```python
from datascience import *
import numpy as np
np.set_printoptions(threshold=50)

import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
students_all = Table().read_table('https://raw.githubusercontent.com/Matthew-HMS/Datascience101/master/110_sdata.csv')
students_all
```

```python
# 刪除技職大學
students_all = students_all.where('體系別', are.equal_to('1 一般'))
students_all
```

```python
# 刪除研究所
students_all = students_all.where('等級別', are.equal_to('B 學士'))
```

```python
# 只留下日間部
students_all = students_all.where('日間∕進修別', are.equal_to('D 日'))
students_all
```

```python
# 將數字中的逗號去除
adjust = students_all.column('學生數')

for i in np.arange(len(adjust)):
    adjust[i] = int(adjust[i].replace(',', ''))

students_all = students_all.with_column('學生數(改)', adjust)
students_all
```

```python
# 將資料型態從<U5轉為int32
change = students_all.column('學生數(改)').astype(np.int32)
students_all = students_all.with_column('學生數(改改)', change)
students_all = students_all.drop('學生數(改)').drop('學生數').relabeled('學生數(改改)', '學生數')
students_all
```

```python
#刪除教師數為0的資料並選出學校名稱、科系名稱、學生數、教師數
students_all = students_all.where('教師數', are.not_equal_to(0))
students = students_all.select('學校名稱', '科系名稱', '學生數', '教師數')
```

以下統計資料皆為台灣普通大學(非技職)日間部且是大學部的學生 ! ! !

## 4. 推算出台灣所有大學校系平均師生比(Bootstrap)

```python
ratio = students.column('學生數') / students.column('教師數')
students = students.with_column('師生比', ratio)
students
```

```python
# 隨機抽取100筆資料
our_sample = students.sample(100, with_replacement = False)
pop_median = percentile(50, students.column('師生比'))
```

```python
percentile(50, our_sample.column('師生比'))
```

```python
def one_bootstrap_median():
    resample = our_sample.sample(100, with_replacement = True)
    return percentile(50, resample.column('師生比'))
```

```python
# 重複抽樣1000次
repetitions = 1000
bstrap_median = make_array()
for i in np.arange(repetitions):
    bstrap_median = np.append(bstrap_median, one_bootstrap_median())
```

```python
resampled_medians = Table().with_column('Bootstrap Sample Median', bstrap_median)
resampled_medians.hist(bins = np.arange(0, 50, 2))

left = percentile(2.5, bstrap_median)
right = percentile(97.5, bstrap_median)

plt.ylim(-0.01, 0.4)
plt.plot([left, right], [0, 0], color='yellow', lw=3, zorder=1);
plt.scatter(pop_median, 0, color = 'red', s=20, zorder=2);
```

## 5. 台灣各縣市師生比差異

```python
students_region = students_all.select('學校名稱', '科系名稱', '學生數', '教師數', '縣市名稱')
students_region
```

```python
students_region_one = students_region.where('縣市名稱', are.equal_to('30 臺北市')).drop('科系名稱').drop('縣市名稱').group('學校名稱', sum)
students_region_one.column('學生數 sum').sum() / students_region_one.column('教師數 sum').sum()
```

```python
region = ['新北市', '宜蘭縣', '桃園市', '苗栗縣', '臺中市', '彰化縣', '南投縣', '嘉義縣', '臺南市', '高雄市', '屏東縣', '臺東縣', '花蓮縣', '基隆市', '新竹市', '嘉義市', '臺北市', '金門縣']
region_ratio = make_array()

for i in region:
    students_region_one = students_region.where('縣市名稱', are.containing(i)).drop('科系名稱').drop('縣市名稱').group('學校名稱', sum)
    ratio = students_region_one.column('學生數 sum').sum() / students_region_one.column('教師數 sum').sum()
    region_ratio = np.append(region_ratio, ratio)

region_ratio
```

```python
region_result = Table().with_columns('縣市名稱', region, '師生比', region_ratio)
region_result.sort('師生比', descending = False).barh('縣市名稱', '師生比')
```

```python
students_region.where('縣市名稱', are.containing('新北')).drop('科系名稱').drop('縣市名稱').group('學校名稱', sum)
```

我們想知道是不是那個縣市私立學校比較多，所以造成師生比較高，所以我們將台灣各縣市的私立學校數量與師生比做了一個散佈圖，但發現並沒有太大關係。

```python
own_gov_ratio = make_array()
for i in region:
    college_region = students_region.where('縣市名稱', are.containing(i)).drop('科系名稱').drop('縣市名稱').group('學校名稱')
    own = college_region.where('學校名稱', are.not_containing('國立')).where('學校名稱', are.not_containing('市立'))
    college_region_ratio = own.num_rows / college_region.num_rows
    own_gov_ratio = np.append(own_gov_ratio, college_region_ratio)
```

```python
own_gov_result = Table().with_columns('縣市名稱', region, '私立比例', own_gov_ratio)
own_gov_result.sort('私立比例', descending = False).barh('縣市名稱', '私立比例')
```

```python
region_and_own = region_result.join('縣市名稱', own_gov_result).sort('師生比', descending = True)
region_and_own.scatter('師生比', '私立比例')
```

以上是我們的分析結果：
師生比是學生人數與教師人數之比，通常會被用來衡量一個學校的教學質量。從中央校內可看出管院及學士班的師生比較高。從全國資管系的師生比可看出，中央資管位於中間，且我們可以推論私立大學普遍較國立大學高，結果如下個分析。從全國各大學師生比可看出，私立平均來說確實比國立還高，且傳統名校四大四中的師生比都偏低，意味著學生有更多的機會得到個別關注和指導。由全國平均師生比來看，中央資管20.333大約就是Bootstrap後的中位數，在全國算是中間值。另外，我們推測師生比較高的縣市是因為私立學校較多，但其實沒有正相關。
大眾通常認為師生比較低，學校評價較高，我們的分析結果確實也呈現相同的結論，所以許多做出大學排名的機構都會將師生比納入評比項目是有原因的。
